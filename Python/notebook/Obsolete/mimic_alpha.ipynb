### mimic_alpha

* Since transparency (alpha) is not allowed in .eps figures, so this routine help to postulate
the color blended with background for mimicking transparency.

```python
import mimic_alpha as ma
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
ax  = fig.add_subplot(111)

x = np.linspace(1,100,10)
y = x*x

alpcolor = ma.colorAlpha_to_rgb('C0',0.5)
ax.plot(x,y,'o',markersize=10,color=alpcolor[0])
```
